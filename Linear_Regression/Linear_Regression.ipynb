```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt

def mean_squared_error(y_true, y_predicted):

	# Calculating the loss or cost
	cost = np.sum((y_true-y_predicted)**2) / (2*len(y_true))
	return cost

# Gradient Descent Function
# Here iterations, learning_rate, stopping_threshold
# are hyperparameters that can be tuned
def give_a_regression_model_step5(X0string,X1string,X2string,X3string,X4string,Ystring,D1,current_weight_0, current_weight_1,current_weight_2,current_weight_3,current_weight_4,current_bias,learning_rate):
		X0 =np.array(D1[X0string])
		X1 =np.array(D1[X1string])
		X2 =np.array(D1[X2string])
		X3 =np.array(D1[X3string])
		X4 =np.array(D1[X4string])
		Y  =np.array(D1[Ystring])
		# Estimating weight and bias using gradient descent
		estimated_weight_0, estimated_weight_1,estimated_weight_2,estimated_weight_3,estimated_weight_4, estimated_bias,current_cost = gradient_descent_step5(X0,X1,X2,X3,X4, Y,current_weight_0,current_weight_1,current_weight_2,current_weight_3,current_weight_4,current_bias,learning_rate)
		print(f"Estimated Weight: {estimated_weight_0}\nEstimated Bias: {estimated_bias}\n total cost: {current_cost}")

		# Making predictions using estimated parameters
		Y_pred = estimated_weight_0*X0 +estimated_weight_1*X1 + estimated_weight_2*X2+estimated_weight_3*X3+estimated_weight_4*X4+estimated_bias

		# Plotting the regression line


def gradient_descent_step5(x0,x1,x2,x3,x4, y,current_weight_0,current_weight_1,current_weight_2,current_weight_3,current_weight_4,current_bias,learning_rate):

	# Initializing weight, bias, learning rate and iterations

	iterations = 5000000
	stopping_threshold = 1e-6
	# learning_rate = 0.0000001
	learning_rate = learning_rate
	n = float(len(x0))

	costs = []
	weights_0 = []
	weights_1 = []
	weights_2 = []
	weights_3 = []
	weights_4 = []
	previous_cost = None

	# Estimation of optimal parameters
	for i in range(iterations):

		# Making predictions
		y_predicted = (current_weight_0 * x0) + (current_weight_1*x1)+(current_weight_2 * x2) + (current_weight_3*x3)+(current_weight_4*x4) + current_bias

		# Calculating the current cost
		current_cost = mean_squared_error(y, y_predicted)

		# If the change in cost is less than or equal to
		# stopping_threshold we stop the gradient descent
		if previous_cost and abs(previous_cost-current_cost)<=stopping_threshold:
			break

		previous_cost = current_cost

		costs.append(current_cost)
		weights_0.append(current_weight_0)
		weights_1.append(current_weight_1)
		weights_2.append(current_weight_2)
		weights_3.append(current_weight_3)
		weights_4.append(current_weight_4)

		# Calculating the gradients
		weight_derivative_0 = -(1/n) * sum(x0 * (y-y_predicted))
		bias_derivative = -(1/n) * sum(y-y_predicted)
		weight_derivative_1 = -(1/n) * sum(x1 * (y-y_predicted))
		weight_derivative_2 = -(1/n) * sum(x2 * (y-y_predicted))
		weight_derivative_3 = -(1/n) * sum(x3 * (y-y_predicted))
		weight_derivative_4 = -(1/n) * sum(x4 * (y-y_predicted))

		# Updating weights and bias
		current_weight_0 = current_weight_0 - (learning_rate * weight_derivative_0)
		current_bias = current_bias - (learning_rate * bias_derivative)
		current_weight_1 = current_weight_1 - (learning_rate * weight_derivative_1)
		current_weight_2 = current_weight_2 - (learning_rate * weight_derivative_2)
		current_weight_3 = current_weight_3 - (learning_rate * weight_derivative_3)
		current_weight_4 = current_weight_4 - (learning_rate * weight_derivative_4)

		# Printing the parameters for each 1000th iteration
		print(f"Iteration {i+1}: Cost {current_cost}, Weight \
		{current_weight_0}, Bias {current_bias}")


	# Visualizing the weights and cost at for all iterations


	return current_weight_0, current_weight_1,current_weight_2,current_weight_3,current_weight_4,current_bias ,current_cost
def gradient_descent_step3(x0,x1, y,current_weight_0,current_weight_1,current_bias,learning_rate):

	# Initializing weight, bias, learning rate and iterations

	iterations = 300000
	stopping_threshold = 1e-6
	# learning_rate = 0.0000001
	learning_rate = learning_rate
	n = float(len(x0))

	costs = []
	weights_0 = []
	weights_1 = []
	previous_cost = None

	# Estimation of optimal parameters
	for i in range(iterations):

		# Making predictions
		y_predicted = (current_weight_0 * x0) + (current_weight_1*x1) + current_bias

		# Calculating the current cost
		current_cost = mean_squared_error(y, y_predicted)

		# If the change in cost is less than or equal to
		# stopping_threshold we stop the gradient descent
		if previous_cost and abs(previous_cost-current_cost)<=stopping_threshold:
			break

		previous_cost = current_cost

		costs.append(current_cost)
		weights_0.append(current_weight_0)
		weights_1.append(current_weight_1)

		# Calculating the gradients
		weight_derivative_0 = -(1/n) * sum(x0 * (y-y_predicted))
		bias_derivative = -(1/n) * sum(y-y_predicted)
		weight_derivative_1 = -(1/n) * sum(x1 * (y-y_predicted))

		# Updating weights and bias
		current_weight_0 = current_weight_0 - (learning_rate * weight_derivative_0)
		current_bias = current_bias - (learning_rate * bias_derivative)
		current_weight_1 = current_weight_1 - (learning_rate * weight_derivative_1)

		# Printing the parameters for each 1000th iteration
		# print(f"Iteration {i+1}: Cost {current_cost}, Weight \
		# {current_weight_0}, Bias {current_bias}")


	# Visualizing the weights and cost at for all iterations


	return current_weight_0, current_weight_1,current_bias ,current_cost

def give_a_regression_model_step3(X0string,X1string,Ystring,D1,current_weight_0, current_weight_1,current_bias,learning_rate):
		X0 =np.array(D1[X0string])
		X1 =np.array(D1[X1string])
		Y  =np.array(D1[Ystring])
		# Estimating weight and bias using gradient descent
		estimated_weight_0, estimated_weight_1, estimated_bias,current_cost = gradient_descent_step3(X0,X1, Y,current_weight_0,current_weight_1,current_bias,learning_rate)
		print(f"Estimated Weight: {estimated_weight_0}\nEstimated Bias: {estimated_bias}\n total cost: {current_cost}")

		# Making predictions using estimated parameters
		Y_pred = estimated_weight_0*X0 +estimated_weight_1*X1 + estimated_bias

		# Plotting the regression line
		fig = plt.figure()
		ax = fig.add_subplot(111, projection='3d')

		ax.scatter(X0, X1, Y_pred)
		x2_range = np.arange(min(X0), max(X1), 0.1)
		x3_range = np.arange(min(X0), max(X1), 0.1)

		#ax.plot(X0, X1, np.array(Y_pred), color='red', linewidth=2)

		ax.set_xlabel('X2')
		ax.set_ylabel('X3')
		ax.set_zlabel('Y')

		plt.show()




def gradient_descent(x, y,current_weight,current_bias,learning_rate):

	# Initializing weight, bias, learning rate and iterations

	iterations = 300
	stopping_threshold = 1e-6

	learning_rate = learning_rate
	n = float(len(x))

	costs = []
	weights = []
	previous_cost = None

	# Estimation of optimal parameters
	for i in range(iterations):

		# Making predictions
		y_predicted = (current_weight * x) + current_bias

		# Calculating the current cost
		current_cost = mean_squared_error(y, y_predicted)

		# If the change in cost is less than or equal to
		# stopping_threshold we stop the gradient descent
		if previous_cost and abs(previous_cost-current_cost)<=stopping_threshold:
			break

		previous_cost = current_cost

		costs.append(current_cost)
		weights.append(current_weight)

		# Calculating the gradients
		weight_derivative = -(1/n) * sum(x * (y-y_predicted))
		bias_derivative = -(1/n) * sum(y-y_predicted)

		# Updating weights and bias
		current_weight = current_weight - (learning_rate * weight_derivative)
		current_bias = current_bias - (learning_rate * bias_derivative)

		# Printing the parameters for each 1000th iteration
		# print(f"Iteration {i+1}: Cost {current_cost}, Weight \
		# {current_weight}, Bias {current_bias}")


	# Visualizing the weights and cost at for all iterations
	plt.figure(figsize = (8,6))
	plt.plot(weights, costs)
	plt.scatter(weights, costs, marker='o', color='red')
	plt.title("Cost vs Weights")
	plt.ylabel("Cost")
	plt.xlabel("Weight")
	plt.show()

	return current_weight, current_bias ,current_cost

def give_a_regression_model(Xstring,Ystring,D1,current_weight,current_bias,learning_rate):
		X =np.array(D1[Xstring])
		Y =np.array(D1[Ystring])
		# Estimating weight and bias using gradient descent
		estimated_weight, estimated_bias,current_cost = gradient_descent(X, Y,current_weight,current_bias,learning_rate)
		print(f"Estimated Weight: {estimated_weight}\nEstimated Bias: {estimated_bias}\n total cost: {current_cost}")

		# Making predictions using estimated parameters
		Y_pred = estimated_weight*X + estimated_bias

		# Plotting the regression line


		plt.plot(X,Y_pred , c ='r',label='eifeif')
		plt.scatter(X,Y_pred,marker='X',c='g')

		plt.show()

def main():
	# Data

		# X =np.array(D1['X2 '])
		# Y =np.array(D1['y'])
		# # Estimating weight and bias using gradient descent
		# estimated_weight, estimated_bias = gradient_descent(X, Y, iterations=200)
		# print(f"Estimated Weight: {estimated_weight}\nEstimated Bias: {estimated_bias}")

		# # Making predictions using estimated parameters
		# Y_pred = estimated_weight*X + estimated_bias

		# # Plotting the regression line


		# plt.plot(X,Y_pred , c ='r',label='eifeif')
		# plt.scatter(X,Y_pred,marker='o',c='g')

		# plt.show()

		plt.style.use('seaborn-poster')

		df = pd.read_csv("Real estate.csv")
		# Load data into a pandas DataFrame


		# Create a subset of the data with the first 200 samples
		D1 = df.head(200)
		D2 = df.head(300)

		# # Plot Y against each of the 5 features separately

		# give_a_regression_model_step3('X2 ','X3 ','y',D1,44,44,44,0.0000001)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,45,45,45,0.1)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,46,46,46,0.1)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,44,44,44,0.01)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,45,45,45,0.01)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,46,46,46,0.01)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,44,44,44,0.001)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,45,45,45,0.001)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,46,46,46,0.001)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,44,44,44,0.0001)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,45,45,45,0.0001)
		# give_a_regression_model_step3('X2 ','X3 ','y',D1,46,46,46,0.0001)
		#give_a_regression_model_step5('X2 ','X3 ','X4 ','X5 ','X6 ','y',D1,46,46,46,46,46,46,0.0000001)

		# give_a_regression_model('X2 ','y',D1,44,44,0.1)
		# give_a_regression_model('X2 ','y',D1,45,45,0.1)
		# give_a_regression_model('X2 ','y',D1,46,46,0.1)
		# give_a_regression_model('X2 ','y',D1,44,44,0.01)
		# give_a_regression_model('X2 ','y',D1,45,45,0.01)
		# give_a_regression_model('X2 ','y',D1,46,46,0.01)
		# give_a_regression_model('X2 ','y',D1,44,44,0.001)
		# give_a_regression_model('X2 ','y',D1,45,45,0.001)
		# give_a_regression_model('X2 ','y',D1,46,46,0.001)
		# give_a_regression_model('X2 ','y',D1,44,44,0.0001)
		# give_a_regression_model('X2 ','y',D1,45,45,0.0001)
		#give_a_regression_model('X2 ','y',D1,46,46,0.0001)

		# give_a_regression_model('X2 ','y',D2,44,44,0.1)
		# give_a_regression_model('X2 ','y',D2,45,45,0.1)
		# give_a_regression_model('X2 ','y',D2,46,46,0.1)
		# give_a_regression_model('X2 ','y',D2,44,44,0.01)
		# give_a_regression_model('X2 ','y',D2,45,45,0.01)
		# give_a_regression_model('X2 ','y',D2,46,46,0.01)
		# give_a_regression_model('X2 ','y',D2,44,44,0.001)
		# give_a_regression_model('X2 ','y',D2,45,45,0.001)
		# give_a_regression_model('X2 ','y',D2,46,46,0.001)
		# give_a_regression_model('X2 ','y',D2,44,44,0.0001)
		# give_a_regression_model('X2 ','y',D2,45,45,0.0001)
	#	give_a_regression_model('X2 ','y',D2,46,46,0.0001)






if __name__=="__main__":
	main()
```
